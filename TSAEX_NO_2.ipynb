***EX.NO 2 VISUALIZING TIME SERIES DATA***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/drive/MyDrive/Electric_Production.csv",parse_dates=["DATE"],index_col="DATE")
print(data.head())
```

```python
plt.figure(figsize=(12,6))
plt.plot(data["IPG2211A2N"],label="TIME SERIES DATA")
plt.xlabel("DATE")
plt.ylabel("IPG2211A2N")
plt.title("TIME SERIES DATA")
plt.legend()
plt.show()
```

```python
monthly_data=data["IPG2211A2N"].resample("M").mean()
plt.figure(figsize=(12, 6))
plt.plot(monthly_data, label="MONTHLY DATA")
plt.title("MONTHLY TRENDS")
plt.xlabel("DATE")
plt.ylabel("Average value")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
data["IPG2211A2N"].hist(bins=30)
plt.title("HISTOGRAM")
plt.xlabel("VALUE")
plt.ylabel("FREQUENCY")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x=data["IPG2211A2N"])
plt.title("BOXPLOT")
plt.xlabel("VALUE")
plt.show()
```
